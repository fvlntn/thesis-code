```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tissue_results_csv = os.path.join('results', 'results_csv.csv')
tissue_csv = pd.read_csv(tissue_results_csv)
```

```python
svbp_csv = tissue_csv[tissue_csv.study == 'SVBP']
map6_csv = tissue_csv[tissue_csv.study == 'MAP6'] 
```

```python
for study_csv in [svbp_csv, map6_csv]:
    for c in study_csv.columns:
        if "volume" in c:
            sns.boxplot(data=study_csv, x="genotype", y=c, hue="method")
            plt.show()
```
