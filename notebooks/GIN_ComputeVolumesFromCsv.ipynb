```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tissue_results_csv = os.path.join('results', 'gin_results', 'volume_results_merged.csv')
tissue_csv = pd.read_csv(tissue_results_csv)
```

```python
svbp_csv = tissue_csv[tissue_csv.study == 'SVBP'][tissue_csv.method != "DL"][tissue_csv.method != "affine"]
map6_csv = tissue_csv[tissue_csv.study == 'MAP6'][tissue_csv.method != "DL"][tissue_csv.method != "affine"]
```

```python
for study_csv in [svbp_csv]:
    for c in study_csv.columns:
        if "volume" in c and not "volume_bs" in c:
            sns.boxplot(data=study_csv, x="genotype", y=c, hue="method")
            plt.show()
```

```python
for study_csv in [map6_csv]:
    for c in study_csv.columns:
        if "volume" in c:
            sns.boxplot(data=study_csv, x="genotype", y=c, hue="method")
            plt.show()
```
